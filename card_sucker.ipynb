```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import re

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode (remove to see browser)
driver = webdriver.Chrome(options=options)
for url in urls:
    driver.get(url)
    
    year = re.search(r'\d{4}|\d{2}', url).group() 
    if year == '17':
        year = '2017'
    os.makedirs(f"cards/{year}", exist_ok=True) 
    
    image_elements = driver.find_elements(By.CSS_SELECTOR, "table.checklist a.thumbnail")
    
    for img in image_elements:
        img_url = img.get_attribute("href")  # Extract image URL
        player_name_elem = img.find_element(By.TAG_NAME, "font")  # Extract player name
        
        if img_url and img_url.endswith(".jpg") and player_name_elem:
            player_name = player_name_elem.text.strip().replace("/", "-")  # Get name & replace invalid chars
            img_name = f"{player_name+year}.jpg"  # Format as filename
            img_path = os.path.join(f"cards/{year}", img_name)
            
            # Download the image
            response = requests.get(img_url)
            with open(img_path, "wb") as file:
                file.write(response.content)
            
            print(f"Downloaded: {img_name}")

# Close the browser
driver.quit()
```

```python
from pathlib import Path
import ollama
import json

folder_path = Path("/Users/jatin/Documents/python/cricket attax/cards/2011")
players = {}  # Initialize dictionary to store extracted data

for file in folder_path.iterdir():
    if file.is_file():
        print(file.name)
        response = ollama.chat(
            model='llama3.2-vision',
            messages=[{
                'role': 'user',
                'content': 'Extract only the numbers from this image (bowling, runs, batting) from left to right without any extra details',
                'images': [file]
            }]
        )
        
        extracted_text = response['message']['content'].strip()
        if extracted_text.endswith('.'):
            extracted_text = extracted_text[:-1]
        if ', ' in extracted_text:
            response_values = extracted_text.split(', ')
        else:
            response_values = extracted_text.split()
        if len(response_values) == 3:
            players[file.name] = {
                'bowling': response_values[0],
                'runs': response_values[1],
                'batting': response_values[2]
            }

# Save results to JSON file
with open("players2011.json", "w") as file:
    json.dump(players, file, indent=4)
```
